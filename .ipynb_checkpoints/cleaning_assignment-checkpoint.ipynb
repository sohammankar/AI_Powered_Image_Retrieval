```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import json
from PIL import Image
from skimage.measure import shannon_entropy
import random
```

### Functions to Detect Jpeg Compression and Blur

```python
def detect_compression(image):
    """Detect JPEG compression artifacts using entropy and blockiness"""
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image
    
    # Calculate entropy (lower entropy often indicates compression)
    entropy = shannon_entropy(gray)
    
    # Simple blockiness detection (JPEG compression creates 8x8 blocks)
    h, w = gray.shape
    block_diff = 0
    count = 0
    
    # Check differences at 8x8 block boundaries
    for i in range(7, h-1, 8):
        for j in range(w-1):
            block_diff += abs(int(gray[i, j]) - int(gray[i+1, j]))
            count += 1
    
    for i in range(h-1):
        for j in range(7, w-1, 8):
            block_diff += abs(int(gray[i, j]) - int(gray[i, j+1]))
            count += 1
    
    blockiness = block_diff / count if count > 0 else 0
    
    # Compression score (higher blockiness / lower entropy = more compression)
    return blockiness / entropy if entropy > 0 else 0

def detect_blur(image):
    """Detect blur using Laplacian variance (lower = more blurry)"""
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image
    
    # Calculate Laplacian and its variance
    laplacian = cv2.Laplacian(gray, cv2.CV_64F)
    return laplacian.var()

def analyze_image_quality(image, blur_threshold=0.0, compress_threshold=0.0):
    detect_dict = {'is_clean': True, 'is_blur': False, 'is_compress': False}
    
    blur_value = detect_blur(image)
    compress_value = detect_compression(image)

    if blur_value < blur_threshold:
        detect_dict['is_blur'] = True
        detect_dict['is_clean'] = False
    if compress_value >= compress_threshold:
        detect_dict['is_compress'] = True
        detect_dict['is_clean'] = False

    return detect_dict
        
        
        
```

## Loading All Image Paths

```python
img_paths = glob.glob('final-corrupt-mini/*jpg', recursive=True)
print('Total Images:', len(img_paths))
```

## Run Cleaning for a Sample Image
Please set the 'blur_threshold' and 'compress_threshold' accordingly to find the right value for cleaning the image

```python
rndn_path = random.sample(img_paths, 1)[0]
print('Image Path:', rndn_path)

org_img = cv2.imread(rndn_path)
detect_dict = analyze_image_quality(org_img, blur_threshold=0.00, compress_threshold=0.00)

print('Result:',detect_dict)
plt.imshow(cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Run Cleaning for the Entire Dataset
Once the correct 'blur_threshold' and 'compress_threshold' are found, replace the thresholds accordingly to apply cleaning on the dataset

```python
result_json = {}
for path in tqdm(img_paths):
    unique_key = path.split('/')[-1].strip('.jpg')
    org_img = cv2.imread(path)
    detect_dict = analyze_image_quality(org_img, blur_threshold=0.00, compress_threshold=0.00)

    result_json[unique_key] = detect_dict
```

## Save the Result Json

```python
save_file_name = 'firstname_lastname.json'

with open(save_file_name, 'w') as f:
    json.dump(result_json, f)
```
