## Scipy

```python
from scipy import linalg

# Coefficient matrix
A = np.array([[3, 2], [1, 4]])
# Right-hand side
b = np.array([6, 8])

# Solve the system
x = linalg.solve(A, b)
print("Solution:", x)
```

## Matplotlib

```python
from matplotlib import pyplot as plt
%matplotlib inline

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-')

# Add titles and labels
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

## Faiss

```python
# Create a dataset of 10 vectors, each with 3 dimensions
data = np.random.random((10, 3)).astype('float32')

# Create a FAISS index
index = faiss.IndexFlatL2(3)  # L2 distance (Euclidean)
index.add(data)

# Query with a random vector
query = np.random.random((1, 3)).astype('float32')
distances, indices = index.search(query, k=2)  # Find 2 nearest neighbors

print("Query vector:", query)
print("Nearest neighbors' indices:", indices)
print("Distances:", distances)
```
